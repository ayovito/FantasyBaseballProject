To get a more holistic profile for each player, we'll need to import data from FanGraphs, drop what columns are redundant and then merge the two datasets together.

```python
# read in 2015 fangraphs data
fangraphs_2015 = pd.read_csv("C:/Users/avitosky/Documents/Baseball Project/fangraphs_2015.csv")
```

```python
# read in 2016 fangraphs data
fangraphs_2016 = pd.read_csv("C:/Users/avitosky/Documents/Baseball Project/fangraphs_2016.csv")
```

```python
# read in 2017 fangraphs data
fangraphs_2017 = pd.read_csv("C:/Users/avitosky/Documents/Baseball Project/fangraphs_2017.csv")
```

```python
# savant data has minimum 75 ABs, fangraphs can only specify minimum PAs so set to 70 but included ABs in data
# need to drop < 75 ABs from each fangraphs dataset
fangraphs_2015 = fangraphs_2015[fangraphs_2015["AB"] >= 75]
```

```python
# same with 2016 data
fangraphs_2016 = fangraphs_2016[fangraphs_2016["AB"] >= 75]
```

```python
# same with 2017 data
fangraphs_2017 = fangraphs_2017[fangraphs_2017["AB"] >= 75]
```

```python
# sort 2015 fangraphs data by name
fangraphs_2015 = fangraphs_2015.sort('Name')
```

```python
# sort 2016 fangraphs data by name
fangraphs_2016 = fangraphs_2016.sort('Name')
```

```python
# sort 2017 fangraphs data by name
fangraphs_2017 = fangraphs_2017.sort('Name')
```

```python
# reset fangraphs 2015 index
fangraphs_2015 = fangraphs_2015.reset_index(drop=True)
```

```python
# reset fangraphs 2016 index
fangraphs_2016 = fangraphs_2016.reset_index(drop=True)
```

```python
# reset fangraphs 2017 index
fangraphs_2017 = fangraphs_2017.reset_index(drop=True)
```

```python
# look at background info, mostly floats and ints with one object (player_name)
fangraphs_2015.info()
print("-------------------------------------")
fangraphs_2016.info()
print("-------------------------------------")
fangraphs_2017.info()
```

```python
# K% and BB% are not objects shown in percentages so we'll need to change those to floats
fangraphs_2015.rename(columns={'BB%':'BB_percentage'}, inplace=True)
fangraphs_2016.rename(columns={'BB%':'BB_percentage'}, inplace=True)
fangraphs_2017.rename(columns={'BB%':'BB_percentage'}, inplace=True)
fangraphs_2015.rename(columns={'K%':'K_percentage'}, inplace=True)
fangraphs_2016.rename(columns={'K%':'K_percentage'}, inplace=True)
fangraphs_2017.rename(columns={'K%':'K_percentage'}, inplace=True)
```

```python
fangraphs_2017.head()
```

```python
# converting 
fangraphs_2015 ['BB_percentage'] = fangraphs_2015['BB_percentage'].replace('%', ' ', regex=True).astype('float')/100
fangraphs_2016 ['BB_percentage'] = fangraphs_2016['BB_percentage'].replace('%', ' ', regex=True).astype('float')/100
fangraphs_2017 ['BB_percentage'] = fangraphs_2017['BB_percentage'].replace('%', ' ', regex=True).astype('float')/100
fangraphs_2015 ['K_percentage'] = fangraphs_2015['K_percentage'].replace('%', ' ', regex=True).astype('float')/100
fangraphs_2016 ['K_percentage'] = fangraphs_2016['K_percentage'].replace('%', ' ', regex=True).astype('float')/100
fangraphs_2017 ['K_percentage'] = fangraphs_2017['K_percentage'].replace('%', ' ', regex=True).astype('float')/100
```

```python
fangraphs_2017.head()
```

```python
# from fangraphs will drop variables Team, G, ISO, BABIP, AVG, SLG, wOBA, Off, Def (both are cumulative stats, not rate stats)
# also drop playerid because it is not consistent from site to site
# drop specified columns from 2015 fangraphs data
fangraphs_2015.drop(['Team', 'AB', 'G', 'ISO', 'BABIP', 'AVG', 'SLG', 'wOBA', 'Off', 'Def', 'playerid'], axis=1, inplace=True)
```

```python
# drop specified columns from 2016 fangraphs data
fangraphs_2016.drop(['Team', 'G', 'AB', 'ISO', 'BABIP', 'AVG', 'SLG', 'wOBA', 'Off', 'Def', 'playerid'], axis=1, inplace=True)
```

```python
# drop specified columns from 2017 fangraphs data
fangraphs_2017.drop(['Team', 'G', 'AB', 'ISO', 'BABIP', 'AVG', 'SLG', 'wOBA', 'Off', 'Def', 'playerid'], axis=1, inplace=True)
```

```python
# change fangraphs 'Name' column to 'player_name' to merge with savant data
fangraphs_2015.rename(columns={'Name':'player_name'}, inplace=True)
```

```python
# change fangraphs 'Name' column to 'player_name' to merge with savant data
fangraphs_2016.rename(columns={'Name':'player_name'}, inplace=True)
```

```python
# change fangraphs 'Name' column to 'player_name' to merge with savant data
fangraphs_2017.rename(columns={'Name':'player_name'}, inplace=True)
```

```python
# merge 2015 fangraphs and savant data on index
merged_2015 = pd.merge(fangraphs_2015, savant_2015, right_index=True, left_index=True)
```

```python
# merge 2016 fangraphs and savant data on index
merged_2016 = pd.merge(fangraphs_2016, savant_2016, right_index=True, left_index=True)
```

```python
# merge 2017 fangraphs and savant data on index
merged_2017 = pd.merge(fangraphs_2017, savant_2017, right_index=True, left_index=True)
```

```python
# check in to see what our merged data looks like
merged_2015.info()
print("-------------------------------------")
merged_2016.info()
print("-------------------------------------")
merged_2017.info()
```
