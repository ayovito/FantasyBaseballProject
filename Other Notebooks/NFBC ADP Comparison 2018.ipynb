```python
# import packages
import pandas as pd
import numpy as np
```

```python
# read in nfbc adp 2017 data
nfbc_adp_2018 = pd.read_csv("C:/Users/avitosky/Documents/Baseball Project/nfbc_adp_2018.csv")
```

```python
nfbc_adp_2018.head()
```

```python
nfbc_adp_2018['Last_Name'], nfbc_adp_2018['First_Name'] = nfbc_adp_2018['Player'].str.split(',',1).str
nfbc_adp_2018['Player_Name'] = nfbc_adp_2018['First_Name'].map(str) + ' ' + nfbc_adp_2018['Last_Name']
```

```python
nfbc_adp_2018.head()
```

```python
nfbc_adp_2018 = nfbc_adp_2018[['Rank', 'Player_Name', 'Team', 'Position(s)', 'ADP', 'Min Pick', 'Max Pick']]
```

```python
nfbc_adp_2018.head()
```

```python
# read in hitting_breakouts_2018_v1 list (top 50%)
hitting_breakouts_2018_v1 = pd.read_csv("C:/Users/avitosky/Documents/Baseball Project/hitting_breakouts_2018_v1.csv", index_col=0)
```

```python
hitting_breakouts_2018_v1.sort_values(by='xwOBA', ascending=False)
```

```python
# strip out any extra white space in Player_Name column before continuing
nfbc_adp_2018['Player_Name'] = nfbc_adp_2018['Player_Name'].str.strip()
hitting_breakouts_2018_v1['Player_Name'] = hitting_breakouts_2018_v1['Player_Name'].str.strip()
```

```python
# before we merge datasets and since this is dataset is different from statcast, let's see if we have any name mismatches
hitting_breakouts_2018_v1[(~hitting_breakouts_2018_v1.Player_Name.isin(nfbc_adp_2018.Player_Name))]
```

```python
# merge datasets on Player_Name
common_2018_v1 = pd.merge(nfbc_adp_2018, hitting_breakouts_2018_v1, on='Player_Name')
```

```python
# show our merged dataset
common_2018_v1.sort_values(by='ADP')
```

In our even more refined breakout list, all of these players could be significant contributors given their hitting profile combined with their expected playing time for the 2018 season.
